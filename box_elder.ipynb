```python
i = 0

while i < len(precincts):
    
    county = 'Box Elder'
    precinct = precincts[i]
    
    k = 1
    x = 0
    
    while x < 25:
        
        vote = df[x][i]
    
        if k % 25 == 0:
            candidate = candidates[k - 1]
            
            k = 1 
        else:
            office = offices[k - 1]
            party = parties[k - 1]
            candidate = candidates[k - 1]
            
            k += 1
            
        if office == 'Democratic for U.S. House 1' or office == 'Republican for U.S. House 1':
            district = '1'
        else: 
            district = ''
            
        d2 = d2.append({
            'county' : county
            ,'precinct' : precinct
            ,'office' : office
            ,'district' : district
            ,'party' : party
            ,'candidate' : candidate
            ,'vote' : vote
        }, ignore_index=True)
        
        x += 1
        
    i += 1
```
